```python
from amopt.pricers.american_fd import american_fd_surface
from amopt.pricers.boundary_extract import extract_boundary_curve
import ipywidgets as widgets
from ipywidgets import Output, VBox, HBox
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

# Used to bind widgets to the plotting function
out = Output()
```

```python
option_type = widgets.Dropdown(options=['call', 'put'], value='call', description='Option Type: ')
K = widgets.IntSlider(value=100, min=0, max=500, step=1, description='ðŸ”¢ Strike Price:')
r = widgets.FloatSlider(value=0.05, max=1, step=0.01, description='ðŸ”¢ Risk-Free Rate:')
q = widgets.FloatSlider(value=0.02, min=0, max=1, step=0.01, description='ðŸ”¢ Dividend Yield')
sigma = widgets.FloatSlider(value=0.2, min=0, max=1, step=0.01, description='ðŸ”¢ Volatility')
T = widgets.IntSlider(value=1, min=1, max=5, step=1, description='ðŸ”¢ Time Period (years)')
```

```python
def compute_surface(_=None):
    with out:
        out.clear_output(wait=True)

        boundary = extract_boundary_curve(option_type.value.lower(),
            K.value,
            r.value,
            q.value,
            sigma.value,
            T.value,
            800,
            800)
        print(boundary[:5], boundary[-5:])
    
        t = np.linspace(0, T.value, len(boundary))
    
        fig, ax = plt.subplots(figsize=(10,6))
        ax.plot(t, boundary, linewidth=2)
        ax.set_title("American Option Free Boundary")
        ax.set_xlabel("Time to Maturity")
        ax.set_ylabel("Boundary Stock Price")
        ax.grid(True)
        
        plt.show()

for w in [option_type, K, r, q, sigma, T]:
    w.observe(compute_surface, names='value')
```

```python
controls = VBox([option_type, K, r, q, sigma, T])
ui = HBox([controls, out])

display(ui)
compute_surface()
```
